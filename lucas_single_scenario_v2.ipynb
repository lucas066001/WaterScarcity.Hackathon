## Policy Definition

Below we define two key policies that regulate water allocation:

1. **Quota Policy**: Determines how much water each actor is allowed to pump
2. **Incentive Policy**: Uses fines and subsidies to influence actor behavior

You can modify these policies to test different regulation strategies.

```python
# Define quota policy - determines how much water each actor is allowed to pump
# This uses the hard_quota policy from the quota_policies module
# In crisis situations, actors with priority below the crisis level get zero quota
def hard_quota(
    self,
    crisis_level: int,
    actors_priority: np.ndarray,
    avg_pump: np.ndarray,
    DOE: float,
    DCR: float
) -> np.ndarray:
    """
    Calcule les quotas d'eau pour chaque acteur en fonction du niveau de crise.
    
    Args:
        crisis_level (int): Niveau de crise actuel (-1 = normal, 0 = alerte, 1+ = crise).
        actors_priority (np.ndarray): Priorité des acteurs (plus la valeur est basse, plus la priorité est basse).
        avg_pump (np.ndarray): Pompage moyen des acteurs.
        DOE (float): Débit optimal écologique.
        DCR (float): Débit critique à ne pas dépasser.
    
    Returns:
        np.ndarray: Quotas d'eau attribués à chaque acteur.
    """
    NORMAL = 0
    ALERT = 0
    CRISIS = 1
    EXTREME_CRISIS = 2

    ESTIMATION_FACTOR = 0.5
    NORMAL_GROWTH_FACTOR = 1.2

    estimated_available_water = 0.0
    quotas = np.zeros_like(avg_pump)

    # Répartition des quotas en fonction du niveau de crise
    if crisis_level == NORMAL:  # Temps normal
        # Les acteurs peuvent pomper jusqu'à 1.9 fois leur moyenne pour maximiser la production
        quotas = avg_pump * NORMAL_GROWTH_FACTOR # Est-ce utile puisque les quotas ne sont pas utilisés en absence de crise

    elif crisis_level == ALERT:  # En période d'alerte available water < DOE
        estimated_available_water = ((DOE + DCR)/2) * (1 + ESTIMATION_FACTOR)  

    elif crisis_level == CRISIS:  # En période d'alerte available water < (DOE + DCR)/2
        estimated_available_water = DCR * (1 + ESTIMATION_FACTOR)

    elif crisis_level == EXTREME_CRISIS:  # En période d'alerte available water < DCR
        estimated_available_water = DCR * (1 - ESTIMATION_FACTOR)

    # Répartition proportionnelle aux priorités et pompages moyens
    priority_sum = np.sum(actors_priority)
    for i in range(len(avg_pump)):
        quotas[i] = (actors_priority[i] / priority_sum) * estimated_available_water
        # Assurez-vous que le quota ne descend pas sous le DCR divisé par le nombre d'acteurs
        quotas[i] = max(quotas[i], DCR / len(avg_pump))

    return quotas

custom_quota = hard_quota

# Define incentive policy - uses fines and subsidies to influence actor behavior
def custom_incentive_policy(
    self,
    actions: np.ndarray,
    actors_priority: np.ndarray,
    avg_incomes: np.ndarray,
    water_pump: np.ndarray,
    avg_pump: np.ndarray,
    is_crisis: np.ndarray,
    water_flows: np.ndarray,
    quota: np.ndarray,
    DOE: float = 15,
    DCR: float = 10
) -> np.ndarray:
    """
    Génère des incitations (bonus ou malus) pour influencer les comportements des acteurs.
    
    Args:
        actors_priority (np.ndarray): Priorité des acteurs.
        avg_incomes (np.ndarray): Revenus moyens des acteurs.
        water_pump (np.ndarray): Quantité d'eau pompée par les acteurs.
        avg_pump (np.ndarray): Pompage moyen des acteurs.
        is_crisis (np.ndarray): Historique des niveaux de crise.
        water_flows (np.ndarray): Débits d'eau disponibles.
        quota (np.ndarray): Quotas attribués aux acteurs.
        DOE (float): Débit optimal écologique.
        DCR (float): Débit critique à ne pas dépasser.
    
    Returns:
        np.ndarray: Tableau des incitations (bonus ou malus) pour chaque acteur.
    """
    incentives = np.zeros_like(water_pump)
    stress_level = is_crisis[-1]  # Niveau de crise actuel
    total_water_used = np.sum(water_pump)

    for i in range(len(water_pump)):
        # Bonus en temps normal si l'acteur dépasser le quota (maximise la production)
        if stress_level == -1 and total_water_used < (water_flows[-1] - DCR):
            if water_pump[i] <= quota[i]:
                incentives[i] = 0  # Neutre si respect du quota
            else:
                # Bonus si dépassement du quota en temps normal
                overuse_ratio = (water_pump[i] - quota[i]) / quota[i]
                incentives[i] = -avg_incomes[i] * overuse_ratio * 2

        # En période de crise, pénaliser les dépassements
        elif stress_level >= 0:
            if water_pump[i] > quota[i]:
                overuse_ratio = (water_pump[i] - quota[i]) / quota[i]
                incentives[i] = avg_incomes[i] * overuse_ratio * (1 + stress_level)  # Malus proportionnel au niveau de crise
            else:
                # Bonus pour les acteurs respectant leur quota
                incentives[i] = -avg_incomes[i] * 0.05

    return np.zeros_like(water_pump)
```

## Simulation Configuration and Execution

Now we'll configure and run the simulation with our defined policies:

```python
 # Configure simulation parameters
yaml_path = 'parameters/scenarios/1.yml'  
params = utils.load_parameters_from_yaml(yaml_path)
params["total_turns"] = 700        # Number of weeks to simulate
params["nb_iterations"] = 20       # Number of simulation runs
params["scarcity"] = "medium"      # Water scarcity level
params["global_forecast_bias"] = -.5       # Bias in water flow forecasts
params["global_forecast_uncertainty"] = .9  # Uncertainty in forecasts

# Initialize and run simulation
simulation = wms.WaterManagementSimulation(**params)

# Apply our custom policies
simulation.incentive_policy = MethodType(custom_incentive_policy, simulation)
simulation.compute_actor_quota = MethodType(custom_quota, simulation)

# Run the simulation
simulation.run_simulation()

# Get final performance scores
ecological_impact, economic_impact = simulation.get_final_scores_scaled()

print(f"Simulation complete. Final scores:")
print(f"Ecological Impact: {ecological_impact:.4f} (lower is better)")
print(f"Economic Impact: {economic_impact:.4f} (higher is better)")

# Prepare for visualization
sns.set(style='whitegrid')
cmap = plt.get_cmap('Dark2')
actors_colors = [cmap(i) for i in range(simulation.nb_actors)]
riverflows_mean = np.mean(simulation.w_riverflows, axis=0)
x_values = range(params["total_turns"])
```

## 1. Water Resource Analysis

First, let's examine the water resource availability throughout the simulation period and how it relates to actor demands.

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 10), constrained_layout=True)

# Plot river flow
gtp.plot_river_flow(axes[0], x_values, riverflows_mean, simulation.actors_demands, simulation.DCR)
axes[0].set_title('River Flow and Actor Demands', fontsize=20)

# Plot remaining water after ecological threshold
gtp.plot_remaining_water(axes[1], x_values, riverflows_mean, simulation.actors_demands, simulation.DCR)
axes[1].set_title('Available Water After Environmental Flow Requirements', fontsize=20)

plt.show()
```

### Analysis of Water Resources

The top graph shows the raw river flow over time, along with each actor's water demand and the ecological threshold (DCR). The total demand line shows the sum of all actor demands.

The bottom graph shows the remaining water available for allocation after respecting the ecological threshold. When this line drops below zero (red shaded areas), it indicates periods where fulfilling all demands would breach the minimum environmental flow requirements.
